Let's start by importing our libraries (seaborn and pandas) and the data we'll be using.  To start, we'll use a dataset called "exercise".  All the datasets we'll use today are included with seaborn.  

Remember to use the play button to run the code in each cell.

```python
import seaborn as sns
import pandas as pd

exercise = sns.load_dataset("exercise")
```

The first thing to do is to have a look at the dataset.

```python
exercise
```

Note that our results are in a column called 'pulse'.  You can check some basic stats for that column by using the describe method on it.

You should see that the average rate is 99.7 bpm, but the results vary from a minimum of 80 bpm to a maximum of 150 bpm.

```python
exercise['diet'].describe()
```

It's always a good idea to see what that data actually looks like.  We can generate a histogram using displot to bucket the data.  Note that you can specify the size of the buckets with 'binwidth'.  Try using smaller and larger buckets to see how the chart changes.

```python
sns.displot(data = exercise, x = "pulse", binwidth=10)
```

Next, let's look for patterns in the data with categorical plotting.  Let's start with just the data.

Notice that we've assigned 'pulse' to the y axis now.

```python
sns.catplot(data = exercise, y = "pulse")
```

We might want to think about what categories to investigate.  Let's see what the pulses looked like for people on different diets.  We do this by assigning 'diet' to the x axis.

```python
sns.catplot(data = exercise, y = 'pulse', x = 'diet')
```

It would make sense for the kind of exercise to also be important, so let's assign 'kind' to the x axis instead.

```python
sns.catplot(data = exercise, y = 'pulse', x = 'kind')
```

It would be nice to see both factors (diet and kind of exercise) at the same time.  We can do that by changing the colour of the dot (or 'hue') according to the diet.

```python
sns.catplot(data=exercise , y = 'pulse', x = 'diet', hue = 'kind')
```
